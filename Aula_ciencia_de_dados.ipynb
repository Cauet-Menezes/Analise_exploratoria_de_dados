```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados = pd.read_csv('perfil-2024.csv', delimiter=';')
```

```python
dados
```

```python
# Padronizando os valores para letras minúsculas e removendo espaços extras
dados['formacao'] = dados['formacao'].str.lower().str.strip()

# Fazendo as substituições
dados['formacao'] = dados['formacao'].replace({
    'tecnologia': 'informática',
    'ciência da computação': 'informática',
    'sistemas de informação': 'informática'
})

# Exibindo os dados atualizados
dados
```

```python
# prompt: gerar com grafico com a coluna de formacao

# Contar a frequência de cada valor na coluna 'formacao'
contagem_formacao = dados['formacao'].value_counts()

# Criar o gráfico de barras
plt.figure(figsize=(10, 6))  # Definir o tamanho da figura
plt.bar(contagem_formacao.index, contagem_formacao.values)
plt.xlabel('Formação')
plt.ylabel('Quantidade')
plt.title('Distribuição de Formação')
plt.xticks(rotation=45, ha='right')  # Rotacionar os rótulos do eixo x para melhor visualização
plt.tight_layout()  # Ajustar o layout para evitar sobreposição de elementos
plt.show()
```
